# Transform data to torch tensor

```python
data_torch = torch.tensor(input_data.values, dtype=torch.float32)
data_torch.size()
```

```python
# Create NN Model

class BinaryClassifier(nn.Module):
    def __init__(self, input_dim):
        super(BinaryClassifier, self).__init__()
        self.layer_1 = nn.Linear(input_dim, 10)
        self.layer_2 = nn.Linear(10, 10)
        self.layer_out = nn.Linear(10, 1) # Single output neuron for binary classification
        self.relu = nn.ReLU()
        # No sigmoid here if using BCEWithLogitsLoss

    def forward(self, inputs):
        x = self.relu(self.layer_1(inputs))
        x = self.relu(self.layer_2(x))
        x = self.layer_out(x)
        return x
```

## Loss Function and Optimizer

```python
model = BinaryClassifier(data_torch) # Initialize your model
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # Set a learning rate
```
